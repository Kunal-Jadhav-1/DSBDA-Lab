```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\91721\Desktop\archive\airquality.csv")
data
```

```python
data
```

```python
ozomean=data["Ozone"].mean()
data["Ozone"].fillna(ozomean,inplace=True)
```

```python
# 1 Scatter Plot
x=data["Temp"]
y=data["Solar.R"]
plt.scatter(x,y,color="red")
plt.ylabel("Solar.R")
plt.xlabel("Temp")
```

```python
# 2 Pie Chart
data['Humidity'].value_counts()
```

```python
labels='high', 'low', 'medium',"low ","high ","medium "
sizes=[data['Humidity'].value_counts()['low'],data['Humidity'].value_counts()['high'],data['Humidity'].value_counts()['medium'],data['Humidity'].value_counts()['medium '],data['Humidity'].value_counts()['low '],data['Humidity'].value_counts()['high ']]
colors=["red","blue","yellow","green","cyan","pink"]
explode=(0,0,0,0.1,0.1,0.1)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%")
plt.title("Pie Chart")
```

```python
# 3 Histogram
x=data['Temp']
plt.hist(x,bins="auto")
plt.title("histogram")
```

```python
# 4 Line Graph
ozone = data['Ozone'].iloc[1:16]
wind = data['Wind'].iloc[1:16]
plt.plot(ozone,label='ozone')
plt.plot(wind,label='wind')
plt.legend()
plt.ylabel("range over 15 days")
plt.xlabel(" days in may")
```

```python
# 5 Bar Plot
ozone = data['Ozone'].iloc[1:16]
days = data['Day'].iloc[1:16]
plt.bar(days,ozone, color ='maroon', width = 0.4)
plt.ylabel("ozone range over 15 days")
plt.xlabel(" days in may")
```

```python
# 6 Pair Plot
# sns.set(style="ticks")
sns.pairplot(data)
```

```python
# 7 Heatmap
data1=data.iloc[1:30,1:6]

ax=sns.heatmap(data1)
# plt.show()
```
